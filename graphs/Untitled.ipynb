```python
import networkx as nx
%matplotlib inline
```

```python
G = nx.Graph()
```

```python
# Add a single Node
G.add_node(1)
```

```python
# Add a list of nodes
G.add_nodes_from([2, 3])
```

```python
nx.draw_networkx(G, node_color="red", node_size=700)
```

```python
G.add_edge(1, 2)
```

```python
nx.draw_networkx(G)
```

```python
e = (2, 3)
G.add_edge(*e)
```

```python
nx.draw_networkx(G)
```

```python
# Getting number of edges and nodes in a graph
print(G.number_of_nodes())
print(G.number_of_edges())
```

#### Get some Properties of a graph

```python
print("Nodes:",G.nodes())
print("Edges", G.edges())
```

```python
## Getting the neighbour of a connected graph
print(G.neighbors(2))
```

```python
G.degree()
```
